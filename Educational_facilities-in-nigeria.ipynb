```python
# Data Cleaning : Check missing values
missing = df.isnull().sum().sort_values(ascending=False)
missing.head(20) 
```

```python
# Handle Missing Values
bool_cols = [
    "improved_water_supply",
    "improved_sanitation",
    "phcn_electricity",
    "chalkboard_each_classroom_yn"
]
df[bool_cols] = df[bool_cols].fillna(False)

num_cols = [
    "num_tchrs_with_nce",
    "num_tchr_full_time",
    "num_students_total",
    "num_classrms_total",
    "num_toilets_total",
    "num_students_male",
    "num_students_female",
    "num_tchrs_male",
    "num_tchrs_female"
]
df[num_cols] = df[num_cols].fillna(0)

cat_cols = [
    "facility_name",
    "facility_type_display",
    "education_type",
    "management",
    "community",
    "ward",
    "unique_lga"
]
df[cat_cols] = df[cat_cols].fillna("Unknown")
```

Justification for filling :
Filling missing values preserves the size and representativeness of the dataset, avoids bias in calculations and visualizations, 
and allows us to accurately calculate metrics like student-teacher ratios, infrastructure access, and facility density.

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
# Fix Datatypes 
bool_cols = [
    "improved_water_supply",
    "improved_sanitation",
    "phcn_electricity",
    "chalkboard_each_classroom_yn"
]

for col in bool_cols:
    df[col] = df[col].astype(bool)
```

```python
df[bool_cols].head(20)
```

```python
from dateutil import parser

def parse_date(date_str):
    try:
        return parser.parse(date_str, dayfirst=False) 
    except (parser.ParserError, TypeError):
        return pd.NaT  

df["date_of_survey"] = df["date_of_survey"].apply(parse_date)
```

```python
df["date_of_survey"].head(20)
```

```python
df["date_of_survey"].isna().sum()
```

```python
num_cols = [
    "num_tchrs_with_nce",
    "num_tchr_full_time",
    "num_students_total",
    "num_classrms_total",
    "num_toilets_total",
    "num_students_male",
    "num_students_female",
    "num_tchrs_male",
    "num_tchrs_female"
]

for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
df[num_cols].head(20)
```

```python
# Drop Unnecessary Columns
df = df.drop(columns=[
    "formhub_photo_id",
    "survey_id",
    "gps",
    "facility_id"
])
```

```python
df.columns
```

```python
# Validate Numeric Columns
for col in num_cols:
    invalid_count = (df[col] < 0).sum()
    print(f"{col} has {invalid_count} negative values")
```

```python
# Create at Least 3 New Features
df["total_students"] = df["num_students_male"] + df["num_students_female"]
df["total_teachers"] = df["num_tchrs_male"] + df["num_tchrs_female"]
```

```python
df.head(5)
```

```python
df["student_teacher_ratio"] = np.where(
    df["total_teachers"] > 0,                
    df["total_students"] / df["total_teachers"],  
    np.nan                                
)

df["student_teacher_ratio"] = df["student_teacher_ratio"].fillna(0)
df["student_teacher_ratio"].head(20)
```

```python
lga_counts = df.groupby("unique_lga")["facility_name"].count().to_dict()
df["facility_density_lga"] = df["unique_lga"].map(lga_counts)
df["facility_density_lga"].head(20)
```

```python
# What are the most common types of educational facilities?
df["facility_type_display"] = df["facility_type_display"].str.strip().str.lower()
df["facility_type_display"] = df["facility_type_display"].replace({
    "primary only": "primary",
    "primary school only": "primary",
    "pre-primary only": "pre-primary",
    "pre-primary and primary": "pre-primary & primary",
    "pre-primary and primary school combined": "pre-primary & primary",
    "junior secondary only": "junior secondary",
    "junior secondary school only": "junior secondary",
    "junior and senior secondary": "junior & senior secondary",
    "junior and senior secondary school combined": "junior & senior secondary",
    "primary and junior secondary": "primary & junior secondary",
    "primary and junior secondary school combined": "primary & junior secondary",
    "primary, junior, and senior secondary": "primary & junior & senior secondary",
    "primary, junior and senior secondary school combined": "primary & junior & senior secondary",
    "information not available / don't know": "unknown",
    "adult, vocational, or technical": "adult/vocational/technical",
    "unknown": "unknown"
})

most_common = df["facility_type_display"].mode()[0]  
count = df["facility_type_display"].value_counts().iloc[0]

print(f"The most common infrastructure is '{most_common}' with {count} schools.")
```

```python
# What percentage of schools have:
# Improved water supply?
# Improved sanitation?
# Electricity?

total_schools = len(df)

water_pct = df["improved_water_supply"].sum() / total_schools * 100
sanitation_pct = df["improved_sanitation"].sum() / total_schools * 100
electricity_pct = df["phcn_electricity"].sum() / total_schools * 100

print(f"Percentage of schools with improved water supply: {water_pct:.2f}%")
print(f"Percentage of schools with improved sanitation: {sanitation_pct:.2f}%")
print(f"Percentage of schools with electricity: {electricity_pct:.2f}%")
```

```python
# Average students per school
avg_students = df["total_students"].mean()

# Average teachers per school
avg_teachers = df["total_teachers"].mean()

print(f"Average students per school: {avg_students:.1f}")
print(f"Average teachers per school: {avg_teachers:.1f}")
```

```python
# Compare public vs private schools in terms of:
# Student population
# Teacher count
# Infrastructure availability

management_stats = df.groupby("management")[[
    "total_students",
    "total_teachers",
    "improved_water_supply",
    "improved_sanitation",
    "phcn_electricity"
]].mean()

management_stats["improved_water_supply"] *= 100
management_stats["improved_sanitation"] *= 100
management_stats["phcn_electricity"] *= 100

management_stats = management_stats.round({
    "total_students": 1,
    "total_teachers": 1,
    "improved_water_supply": 2,
    "improved_sanitation": 2,
    "phcn_electricity": 2
})

management_stats
```

```python
# Identify states or LGAs with:
# The highest number of schools
# The lowest access to electricity and sanitation

lga_summary = df.groupby("unique_lga").agg(
    total_schools = ("facility_name", "count"),
    electricity_pct = ("phcn_electricity", "mean"),
    sanitation_pct = ("improved_sanitation", "mean")
)

lga_summary["electricity_pct"] *= 100
lga_summary["sanitation_pct"] *= 100

lga_summary = lga_summary.round({
    "electricity_pct": 2,
    "sanitation_pct": 2
})

lowest_electricity = lga_summary.sort_values(by="electricity_pct").head(10)
lowest_sanitation = lga_summary.sort_values(by="sanitation_pct").head(10)

print("Top 10 LGAs with lowest electricity access:")
print(lowest_electricity)

print("\nTop 10 LGAs with lowest sanitation access:")
print(lowest_sanitation)
```

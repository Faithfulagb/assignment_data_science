```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("food_coded.csv")
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
missing_values = df.isnull().sum().sort_values(ascending=False)
missing_values.head(5)
```

```python
df.dtypes
```

```python
numeric_columns = ['GPA', 'weight', 'calories_chicken', 'calories_day', 'calories_scone', 'fruit_day']

for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df[numeric_columns].dtypes
```

```python
df[numeric_columns].isnull().sum()
```

```python
for col in numeric_columns:
    df[col] = df[col].fillna(df[col].median())
```

```python
df[numeric_columns].isnull().sum()
```

```python
df['Gender'] = df['Gender'].map({1.0: 'Female', 2.0: 'Male'})
df['Gender'].unique()
```

```python
df['Gender']
```

```python
cuisine_map = {1.0: 'Italian', 2.0: 'Mexican', 3.0: 'Chinese'}
df['cuisine'] = df['cuisine'].map(cuisine_map)

exercise_map = {1.0: 'Daily', 2.0: 'Sometimes', 3.0: 'Never'}
df['exercise'] = df['exercise'].map(exercise_map)
```

```python
df['cuisine']
```

```python
text_columns = ['comfort_food_reasons', 'comfort_food', 'diet_current']
for col in text_columns:
    df[col] = df[col].astype(str).str.strip().str.title()
```

```python
df[text_columns].head(5)
```

```python
numeric_columns = ['GPA', 'weight', 'calories_chicken', 'calories_day', 'calories_scone', 'fruit_day']

df = df[df['GPA'].between(0, 4)]
df = df[df['weight'].between(50, 400)]
df = df[df['calories_chicken'].between(0, 5000)]
df = df[df['calories_scone'].between(0, 5000)]

df[numeric_columns].describe()
```

```python
def bmi_category(weight):
    if weight < 120:
        return 'Underweight'
    elif 120 <= weight <= 180:
        return 'Normal'
    else:
        return 'Overweight'

df['BMI_Category'] = df['weight'].apply(bmi_category)
df[['weight', 'BMI_Category']].head(10)
```

```python
gender_group = df.groupby('Gender')[['GPA', 'fruit_day', 'calories_day']].mean()
gender_group
```

```python
avg_gpa = df['GPA'].mean()
df['GPA_Level'] = df['GPA'].apply(lambda x: 'Above Average' if x > avg_gpa else 'Below Average')
df[['GPA', 'GPA_Level']].head(10)
```

```python
all_foods = df['comfort_food'].dropna().str.split(',').explode().str.strip().str.title()

most_common_food = all_foods.value_counts().head(10)
most_common_food
```

```python
exercise_gpa = df.groupby('exercise')['GPA'].mean().sort_values(ascending=False)
exercise_gpa
```

```python
df['coffee_type'] = df['coffee'].map({1: 'Frappuccino', 2: 'Espresso'})
coffee_gender = df.groupby('Gender')['coffee_type'].value_counts()
coffee_gender
```

```python
cuisine_counts = df['cuisine'].value_counts()
cuisine_counts
```

```python
eat_out_gpa = df.groupby('eating_out')['GPA'].mean()
eat_out_gpa
```

```python
diet_group = df.groupby('diet_current')[['GPA', 'calories_day', 'fruit_day']].mean()
diet_group.head(10)
```

```python
healthy_counts = df[df['healthy_feeling'] >= 4].groupby('parents_cook')['healthy_feeling'].count()
healthy_counts
```

```python
comfort_reason_gpa = df.groupby('comfort_food_reasons_coded')['GPA'].mean().sort_values(ascending=False)
comfort_reason_gpa
```

```python
calories_gender = df.groupby('Gender')[['calories_chicken', 'calories_scone']].mean()
calories_gender
```

```python
# Food habits and nutrition patterns

# Comfort foods: Ice cream (42), pizza (37), and chocolate (25) are the most popular among students. This shows students favor high-calorie, indulgent foods.

# Cuisine preferences: Italian dominates (86 students), with Mexican (13) and Chinese (3) much less popular.

# Coffee consumption: Female students prefer espresso (58) over Frappuccino (18), and males similarly prefer espresso (36) over Frappuccino (13). 

# Eating out vs diet: Students who eat out more (eating_out = 3–5) have slightly lower GPAs (3.33–3.40) than those eating out less (eating_out = 1, GPA 3.48), suggesting a possible negative correlation between frequent fast-food consumption and diet quality.


# Academic performance (GPA)

# Exercise and GPA: Students who exercise daily have the highest GPA (3.43), followed by sometimes (3.42), then never (3.25). This shows a positive correlation between regular exercise and academic performance.

# Comfort food reasons: Students whose comfort food reasons are coded as 6.0 or 8.0 have the highest GPA (3.9 and 3.7), while reason 9.0 correlates with the lowest GPA (2.88).

# Diet type and GPA: Some structured or healthy diets (“2 meals a day, fairly balanced” → GPA 3.7; “Balanced” → GPA 3.7) correspond with higher GPAs, showing diet quality may relate to performance.


#  Relationship between food choices and lifestyle

# Parents cook & health perception: Students whose parents cook regularly (1) report feeling healthy (53 students) more often than those whose parents don’t (29) or are in category 3 (8). Home-cooked meals appear to promote healthier lifestyle habits.

# Calories perception by gender: Female students estimate slightly more calories for chicken (587 vs 561) but fewer for scones (476 vs 549), indicating slight gender differences in food perception.
```

```python
# Recommendations for improving students’ eating habits

# Encourage home-cooked meals or balanced campus meal options, since students with parents who cook regularly report feeling healthier.

# Promote regular exercise, as it’s linked to higher GPA and likely better overall health.

# Offer healthier comfort food alternatives, since indulgent foods like ice cream, pizza, and chocolate are extremely popular.

# Provide nutrition education workshops focused on diet balance, portion sizes, and eating out alternatives.
```
